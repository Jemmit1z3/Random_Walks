This script simulates the **final positions** of many 1D random walkers after a fixed number of steps.

What You'll Do:
- Simulate 1000 walkers, each taking 100 steps.
- Record where each walker ends up.
- Plot a histogram showing how final positions are distributed.
- Print the histogram bin width for interpretation.

This experiment explores how randomness accumulates and reveals a classic bell-shaped patternâ€”even though each individual path is unpredictable.

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
num_walkers = 1000
num_steps = 100
```

```python
# TOD0: Simulate Random Walks
# Get the steps
steps = np.random.choice([-1, 1], size=(num_walkers, num_steps))
# Calculate positions
positions = np.cumsum(steps, axis=1)
# Get the final positions
final_positions=positions[:,-1]
```

```python
# TODO: Plot the histogram
plt.figure(figsize=(8, 4))
# Plot the histogram of final positions
counts, bin_edges, patches =plt.hist(final_positions, bins=30)
bin_width = bin_edges[1] - bin_edges[0]
print(f"Bin Width (interval size): {bin_width}")
plt.title("Distribution of Final Positions (After 100 Steps)")
plt.xlabel("Final Position")
plt.ylabel("Number of Walkers")
plt.grid(True)
plt.show()
```

This simulation demonstrates a key result of probability theory:
the **distribution of outcomes in a random walk approaches a normal (bell-shaped) curve**.

Key insights:
- Even though each walker moves randomly, the collective result forms a **predictable pattern**.
- Most walkers stay near the origin, while fewer reach the extremes.
- The histogramâ€™s shape reflects the **Central Limit Theorem**, which says that sums of random steps tend toward a normal distribution.

This experiment helps explain why random walk models are useful in:
