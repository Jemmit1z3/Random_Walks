This script wraps up everything you've learned about 1D random walks
through three final challenge problems that combine visualization, recurrence, and return statistics.

What You'll Do:
1. Visualize multiple random walkers and compare them to the mean and standard deviation.
2. Check whether walkers return to the origin at least once.
3. Count how many times a walker returns to the origin in 1000 steps.

Together, these challenges reinforce the key patterns and statistical behaviors that emerge in random systems.

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Challenge 1: Plotting Individual Walkers vs. Mean and Spread

# Parameters
num_walkers = 500
num_steps = 100

steps=np.random.choice([-1,1],size=(num_walkers,num_steps))# Get steps
positions=np.cumsum(steps, axis=1)# Calculate positions

mean=np.mean(positions,axis=0)# Calculate mean
std=np.std(positions, axis=0)# Calculate standard deviation

# Plot a few individual paths
plt.figure(figsize=(10, 5))
for i in range(5):
  plt.plot(positions[i])# Plot positions

plt.plot(mean)# Overlay mean and +-1 standard deviation
# Plot mean
plt.fill_between(range(num_steps), mean-std, mean+std, color='gray', alpha=0.3, label='+-1 Standard Deviation')
plt.xlabel("Step")
plt.ylabel("Position")
plt.title("Individual Walkers vs. Mean and Spread")
plt.legend()
plt.show()
```

```python
# Challenge 2: Checking if walkers return to the origin at least once

# Parameters
num_walkers = 1000
num_steps = 1000
returns_to_origin=0

for i in range(num_walkers):# Loop over number of walkers
  step=np.random.choice([-1,1],size=(1,num_steps))# Get steps for each walker
  positions=np.cumsum(step) # Calculate positions
  if 0 in positions:# Check if 0 is in positions
    returns_to_origin += 1

print(f"Out of {num_walkers} walkers, {returns_to_origin} returned to the origin at least once.\n")
```

```python
# Challenge 3: Counting how many times a walker returns to origin

# Parameters
position = 0
returns = 0

for _ in range(1000):
  step=np.random.choice([-1,1])# Get steps
  position+=step# Calculate positions
  if position==0:# Check if position is 0
    returns += 1

print(f"The walker returned to the origin {returns} times in 1000 steps.\n")
```

This wrap-up brings together several key concepts:

- Even though each walker moves randomly, their collective behavior follows **predictable patterns**.
- The **mean position** of unbiased walks stays around zero, while the **standard deviation grows** like √n.
- **Recurrence** is a defining feature of 1D random walks — the walker returns to the origin almost surely.
- You explored both **whether** and **how often** returns happen — key ideas in both probability and real-world modeling.
