This script tracks the **average position** and the **spread (standard deviation)** of many random walkers over time.

What You'll Do:
- Simulate 1000 random walkers taking 100 steps.
- At each step, calculate:
  - The **mean** position of all walkers.
  - The **standard deviation** of their positions (how spread out they are).
- Plot both metrics to visualize how they evolve over time.

This helps illustrate the difference between expected position and variability â€” and reinforces that random systems can have predictable structure in aggregate.

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
num_walkers = 1000
num_steps = 100
```

```python
# TODO: Simulate Random Walks
step = np.random.choice([-1,1], size=(num_walkers, num_steps))# Get the steps
sum_positions=np.cumsum(step, axis=1)# Calculate the positions
```

```python
# TODO: Track statistics
mean=np.mean(sum_positions,axis=0)# Calculate the means
std=np.std(sum_positions,axis=0)# Calculate the standard deviations
```

```python
# TODO: Plot the statistics
plt.figure(figsize=(8, 4))
plt.plot(range(num_steps),mean,label="Mean")# Plot the means
plt.plot(range(num_steps), std, label= "Std")# Plot the standard deviations
plt.title("Mean and Standard Deviation Over Time")
plt.xlabel("Step Number")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()

#plt.plot(std_positions)
```

This simulation reveals two crucial aspects of random walk behavior:

1. **Mean Position**:
   - The average position across all walkers remains close to 0.
   - This reflects the fact that the walk is **unbiased**: no preferred direction.

2. **Standard Deviation**:
   - The spread of walker positions increases over time.
   - It grows roughly as (sqrt{n}), which is a hallmark of diffusion and Brownian motion.

Together, these measures show that:
- Even though the mean doesnâ€™t drift, **individual paths diverge**.
- Random walks are unpredictable at the individual level, but **statistically predictable in groups**.

This sets the stage for deeper topics like **variance growth**, **scaling laws**, and eventually **2D/3D diffusion models**.
